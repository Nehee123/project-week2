# MURDER

```python
murder.head()
```

```python
murder.isna().sum()
```

```python
murder.fillna(murder.mean(),inplace=True)
```

```python
murder.isna().sum()
```

```python
murderg=murder.groupby(['Year','Sub_Group_Name'])['Victims_Total'].sum().reset_index()
murderg
```

```python
murderg=murderg[murderg['Sub_Group_Name']!='3. Total']
murderg
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(14,10))
ax=sns.barplot(x='Year',y='Victims_Total',hue='Sub_Group_Name',data=murderg)
plt.title('Gender distribution of Victims',size=30)
```

```python
murders=murder[murder['Sub_Group_Name']=='3. Total']
murders
```

```python
murders=murders.groupby(['Area_Name'])['Victims_Total'].sum().sort_values(ascending=False).reset_index()

print("Top five states with highest murder cases")
murders.head()
```

```python
print("Top five states with lowest murder cases")
murders.tail()
```
